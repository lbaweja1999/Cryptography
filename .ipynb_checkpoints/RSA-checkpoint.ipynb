```python
import sympy,math,random

p=int(input("Enter p (should be a prime number):"))
q=int(input("Enter q (should be a prime number):"))
# b=int(input("Enter e:"))
plaintext=int(input("Enter plaintext:"))

n=p*q

print()

print("p is:",p)
print("q is:",q)
print("n is:",n)

def eulers_totient():
    phi_n=0
    if(sympy.isprime(p) and sympy.isprime(q)):
        phi_n=(p-1)*(q-1)
    
    else:
        print("p and q should be prime numbers ")
        
    return phi_n

a=eulers_totient()
print("Phi("+str(n)+") is:"+str(a))

def get_e():
    list_e=[]
    a=eulers_totient()
    for i in range(1,a):
        if(math.gcd(i,a)==1):
            list_e.append(i)
        
    e=random.choice(list_e)
    return e
b=get_e()
print("e is:",b)

public_key=(b,n)


def mul_inverse(b,a) : 
    b=b%a 
    for inv_e in range(1,a) : 
        if((b*inv_e)%a==1) : 
            return inv_e 
    return 1

d=mul_inverse(b,a)%a
private_key=(d,n)
print("d is:",d)

print("Public key is:",public_key)
print("Private key is:",private_key)

def cipher_text():
    if(plaintext<n):
        c=pow(plaintext,b)%n
    return c
c=cipher_text()
print("Cipher text is:",c)

def plain_text():
    plaintext=pow(c,d)%n
    return plaintext
ptext=plain_text()
print("Plain text is:",ptext)
```
